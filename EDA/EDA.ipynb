```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np 
from pymongo import MongoClient
import warnings
warnings.filterwarnings("ignore")
```

```python
def read_data(mongo_uri):
    client=MongoClient(mongo_uri)
    db=client['aqi_data']
    collection=db['karachi_aqi_etl']
    data=pd.DataFrame(list(collection.find()))
    return data

data=read_data("mongodb+srv://hasnainhissam56_db_user:Hasnain1234#@cluster0.idmsuf2.mongodb.net/?appName=Cluster0")
data.head()
```
